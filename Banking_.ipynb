```python
from transformers import pipeline
qa_pipeline = pipeline("question-answering")
```

```python
context = """
Federal Deposit Insurance Corporation. It insures deposits in banks up to a certain amount, protecting depositors' money in case of bank failure.
 A debit card is linked directly to your bank account, and purchases are deducted immediately. A credit card allows you to borrow money to make purchases, which you then repay later (with interest if not paid in full by the due date).
Annual Percentage Rate. It's the annual interest rate charged on a loan or credit card. It's important for comparing loan offers and understanding the true cost of borrowing.

"""
```

```python
questions = [
    " What does FDIC stand for,and what doesit do?",
    "What is the difference between a debit card and a credit card?",
    "What is an APR, and why is it important?"
]

for q in questions:
    result = qa_pipeline(question=q, context=context)
    print(f"Q: {q}\nA: {result['answer']}\n")
```
